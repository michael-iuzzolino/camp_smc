```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
from modules.DynamicsProfiles import *
from sklearn.utils import shuffle

%matplotlib inline
```

```python
def get_data():
    data = {}
    with h5py.File("data/model_db.h5", "r") as infile:
        label_names = infile.attrs["label_names"]
        for model_name, model_members in infile.items():
            data[model_name] = [np.array(model_data) for model_data in model_members.values()]
        
    return data, label_names
```

```python
data, label_names = get_data()

plt.plot(data["Altostratus"][0])
plt.title("Altostratus")
```

```python
classes = data.keys()
num_classes = len(classes)
labels = np.identity(num_classes)
label_names = []
X_data = []
y_data = []
for i, (model_name, model_data) in enumerate(data.items()):
    label_names.append(model_name)
    intensities = []
    for intensity in model_data:
        X_data.append(intensity)
        y_data.append(labels[i])
        
X_data = np.array(X_data)
y_data = np.array(y_data)
```

```python
X_shuff, y_shuff = shuffle(X_data, y_data)
```

# AUTOENCODER TEST

```python
import tensorflow as tf
from tensorflow.contrib.layers import fully_connected
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import sys
```

### Architecture

```python
n_inputs = X_data.shape[1]
n_hidden_1 = 100
n_embbed_layer = 100
n_hidden_3 = n_hidden_1
n_outputs = n_inputs

    
X = tf.placeholder(tf.float32, [None, n_inputs])
hidden_1 = fully_connected(X, n_hidden_1, activation_fn=tf.nn.tanh)
embed_layer = fully_connected(hidden_1, n_embbed_layer, activation_fn=tf.nn.tanh)
hidden_3 = fully_connected(embed_layer, n_hidden_3, activation_fn=tf.nn.tanh)
logits = fully_connected(hidden_3, n_outputs, activation_fn=None)
output = tf.nn.sigmoid(logits)
```

### Loss, Optimizer, Init

```python
loss = tf.reduce_sum(tf.nn.sigmoid_cross_entropy_with_logits(labels=X, logits=logits))
optimizer = tf.train.AdamOptimizer()
training_op = optimizer.minimize(loss)

init = tf.global_variables_initializer()
```

### Setup data

```python
X_train, X_test, y_train, y_test = train_test_split(X_shuff, y_shuff)

SCALES = [MinMaxScaler, StandardScaler]
scaler = SCALES[0](feature_range=(0, 1))
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)

# Test set will be randomly chopped up
chop_up = False
if chop_up:
    domain_size = 25
    new_X_test_scaled = []
    for i, X_t in enumerate(X_test_scaled):
        start_index = np.random.randint(0, 50)
        new_Xt = X_t[start_index:start_index+domain_size]
        new_Xt = np.r_[new_Xt, np.zeros(75)]
        new_X_test_scaled.append(new_Xt)

    X_test_scaled = np.array(new_X_test_scaled)
```

### TRAIN

```python
n_epochs = int(1e4)

train_feed_dict = {X: X_train_scaled}
test_feed_dict = {X: X_test_scaled}

plt.ion()
with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        sess.run(training_op, feed_dict=train_feed_dict)
        sys.stdout.write("\r {:5.2f}% complete".format(epoch / float(n_epochs-1) * 100))
        sys.stdout.flush()
    
    log_val, out_val = sess.run([logits, output], feed_dict=train_feed_dict)
    test_out_val = logits.eval(feed_dict=test_feed_dict)
    
    train_embed_val = embed_layer.eval(feed_dict=train_feed_dict)
    test_embed_val = embed_layer.eval(feed_dict=test_feed_dict)
```

```python
def compare_outputs(autoencoder_out, ground_truth, set_type):
    random_index = np.random.randint(len(autoencoder_out))
    fig, ax = plt.subplots(2)
    ax[0].set_title("{} Set - Autoencoder Output".format(set_type))
    ax[0].plot(autoencoder_out[random_index], 'r')
    ax[1].set_title("{} Set - Ground Truth".format(set_type))
    ax[1].plot(ground_truth[random_index])
    plt.tight_layout()
```

```python
compare_outputs(log_val, X_train, "Training")
```

```python
compare_outputs(test_out_val, X_test, "Test")
```

# Compare

### Randomly pick target from train set

```python
compare_index = np.random.randint(len(test_embed_val))

target = test_embed_val[compare_index]

target_label = label_names[np.argmax(y_test[compare_index])]
print("Target: {}".format(target_label))

plt.plot(X_train[compare_index])
plt.title("Target: {}".format(target_label))
```

### Find Max Match

```python
max_val = 0
max_match = None
for i, embed in enumerate(train_embed_val):
    corr = np.dot(target, embed)
    if corr > max_val:
        max_val = corr
        max_match = i
```

### Plot Match

```python
match_label = label_names[np.argmax(y_train[max_match])]
print("Match: {}".format(match_label))

plt.plot(X_train[max_match])
plt.title("Match: {}".format(match_label))
```
