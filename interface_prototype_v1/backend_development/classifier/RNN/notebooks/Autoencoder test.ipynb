```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
from modules.DynamicsProfiles import *
%matplotlib inline
```

```python
MODELS = [Cirrus, Stratus, Cumulus, Nimbostratus, Cumulonimbus, Altostratus]

NOISE = True
NOISE_TYPE = "uniform"
startPose = [40, 40]

feed_dict = {
    "startPose"  : startPose,
    "noise"      : NOISE,
    "noise_type" : NOISE_TYPE
}

all_profiles = [model(**feed_dict) for model in MODELS]
```

```python
full_data = {}
for model in all_profiles:
    model_name = model.name
    model_intensity = model.intensity
    full_data[model_name] = model_intensity
    plt.plot(model_intensity)
    plt.title("Class: {}".format(model_name))
    plt.show()
```

```python
classes = full_data.keys()
num_classes = len(classes)
X_data = np.array(full_data.values())
y_data = np.identity(num_classes)
```

```python
import tensorflow as tf
from tensorflow.contrib.layers import fully_connected
from sklearn.model_selection import train_test_split
import sys
```

```python
n_inputs = X_data.shape[1]
n_hidden_1 = 100
n_hidden_2 = 50
n_hidden_3 = n_hidden_1
n_outputs = n_inputs

    
X = tf.placeholder(tf.float32, [None, n_inputs])
hidden_1 = fully_connected(X, n_hidden_1, activation_fn=tf.nn.relu)
hidden_2 = fully_connected(hidden_1, n_hidden_2, activation_fn=tf.nn.relu)
hidden_3 = fully_connected(hidden_2, n_hidden_3, activation_fn=tf.nn.relu)
logits = fully_connected(hidden_3, n_outputs, activation_fn=None)
```

```python
loss = tf.reduce_sum(tf.nn.sigmoid_cross_entropy_with_logits(labels=X, logits=logits))
optimizer = tf.train.AdamOptimizer()
training_op = optimizer.minimize(loss)

init = tf.global_variables_initializer()
```

```python
X_train = X_data[3]
```

```python
n_epochs = 50000

plt.ion()
with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        sess.run(training_op, feed_dict={X: X_train.reshape(1, 100)})
#         sys.stdout.write("\r {:5.2f}% complete".format(epoch / float(n_epochs) * 100))
#         sys.stdout.flush()
        logits_val = logits.eval(feed_dict={X: X_train.reshape(1, 100)})
        logits_val = logits_val.reshape(100, 1)

        plt.plot(logits_val, 'r')
        plt.pause(0.05)
        plt.cla()
    
```
